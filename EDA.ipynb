**Capstone Project Title**

**AeroSky Analytics: A Data-Driven Atmospheric Suitability System for Aviation Operations and Astronomical Observations**

**Research Problem**

Airport meteorologists and aviation personnel spend significant time analyzing weather conditions to support safe flight operations. At the same time, astronomers must evaluate atmospheric conditions before conducting observations.

Although both fields rely on similar atmospheric parameters such as visibility, cloud cover, humidity, and wind speed, there is no simple data-driven tool that combines these factors into a single assessment that can assist decision-making.

This project aims to develop an Atmospheric Suitability Index (ASI) that automatically evaluates weather conditions and provides recommendations for both aviation and astronomy activities.

**Main Objective**

To develop a data analytics decision-support system that analyzes atmospheric conditions and assists airport meteorologists by automatically evaluating weather suitability for:

1. Aviation operations
2. Astronomical observations

**Specific Objectives**

1. Collect and prepare atmospheric and astronomical datasets.
2. Explore relationships between weather variables.
3. Identify factors that most influence observation and operational quality.
4. Develop an Atmospheric Suitability Index (ASI).
5. Create an interactive dashboard for airport meteorologists and astronomers.
6. Provide recommendations that can support meteorologists, air traffic controllers, and astronomers.
